```python
# Função para criar conexão no banco
import psycopg2
def conecta_db():
  con = psycopg2.connect(host='localhost', 
                         database='dashboard',
                         user='postgres', 
                         password='dantas')
  return con
```

Função para criar tabela no banco

```python
# Função para criar tabela no banco
def criar_db(sql):
  con = conecta_db()
  cur = con.cursor()
  cur.execute(sql)
  con.commit()
  con.close()
```

Criando tabelas

```python
sql_occurrence_details = '''
CREATE TABLE OccurrenceDetails(
  gbifID                BIGINT NOT NULL PRIMARY KEY,
  accessRights          VARCHAR(240) NOT NULL,
  license               VARCHAR(240) NOT NULL,
  rightsHolder          VARCHAR(240) NOT NULL,
  institutionCode       VARCHAR(240) NOT NULL,
  collectionCode        VARCHAR(240) NOT NULL,
  basisOfRecord         VARCHAR(240) NOT NULL,
  occurrenceID          VARCHAR(240) NOT NULL,
  catalogNumber         VARCHAR(240) NOT NULL,
  recordedBy            VARCHAR(240),
  occurrenceStatus      VARCHAR(240) NOT NULL,
  preparations          VARCHAR(240),
  eventDate             VARCHAR(240),
  Tyear                VARCHAR(240),
  TMonth                VARCHAR(240),
  TDay                 VARCHAR(240),
  eventRemarks          VARCHAR(240),
  higherGeography       VARCHAR(240),
  continent             VARCHAR(240),
  countryCode           VARCHAR(240),
  stateProvince         VARCHAR(240),
  county                VARCHAR(240),
  locality              VARCHAR(240),
  decimalLatitude       NUMERIC(15,10),
  decimalLongitude      NUMERIC(15,10),
  typeStatus            VARCHAR(240)
);
'''

# SQL para criar a tabela TaxonomicInformation
sql_taxonomic_information = '''
CREATE TABLE TaxonomicInformation(
  gbifID                BIGINT NOT NULL PRIMARY KEY,
  taxonKey              INTEGER NOT NULL,
  acceptedTaxonKey      NUMERIC(15,2),
  kingdomKey            BIT NOT NULL,
  phylumKey             NUMERIC(15,2),
  classKey              NUMERIC(15,2),
  orderKey              NUMERIC(15,2),
  familyKey             NUMERIC(15,2),
  genusKey              NUMERIC(15,2),
  speciesKey            NUMERIC(15,2),
  species               VARCHAR(240),
  acceptedScientificName VARCHAR(240),
  verbatimScientificName VARCHAR(240),
  FOREIGN KEY (gbifID) REFERENCES OccurrenceDetails(gbifID)
);
'''

# SQL para criar a tabela SpeciesInformation
sql_species_information = '''
CREATE TABLE SpeciesInformation(
 gbifID                BIGINT NOT NULL PRIMARY KEY,
 scientificName        VARCHAR(240) NOT NULL,
 kingdom               VARCHAR(240) NOT NULL,
 phylum                VARCHAR(240),
 Tclass                VARCHAR(240),
 TOrder                VARCHAR(240),
 Tfamily               VARCHAR(240),
 genus                VARCHAR(240),
 genericName           VARCHAR(240),
 specificEpithet       VARCHAR(240),
 infraspecificEpithet  VARCHAR(240),
 taxonRank              VARCHAR(240) NOT NULL,
 taxonomicStatus       VARCHAR(240),
 FOREIGN KEY (gbifID) REFERENCES OccurrenceDetails(gbifID)
);
'''

# SQL para criar a tabela DatasetInformation
sql_dataset_information = '''
CREATE TABLE DatasetInformation(
 gbifID              BIGINT NOT NULL,
 datasetKey          VARCHAR(240) NOT NULL,
 publishingCountry   VARCHAR(240) NOT NULL,
 lastInterpreted     VARCHAR(240) NOT NULL,
 issue               VARCHAR(240) NOT NULL,
 hasCoordinate       VARCHAR(240) NOT NULL,
 hasGeospatialIssues VARCHAR(240) NOT NULL,
 FOREIGN KEY (gbifID) REFERENCES OccurrenceDetails(gbifID)
);
'''

# SQL para criar a tabela ProtocolAndIUCNInformation
sql_protocol_and_iucn_information = '''
CREATE TABLE ProtocolAndIUCNInformation(
 gbifID              BIGINT NOT NULL,
 protocol            VARCHAR(240) NOT NULL,
 lastParsed          VARCHAR(240) NOT NULL,
 lastCrawled         VARCHAR(240) NOT NULL,
 repatriated         VARCHAR(240),
 level0Gid           VARCHAR(240),
 level0Name          VARCHAR(240),
 level1Gid           VARCHAR(240),
 level1Name          VARCHAR(240),
 level2Gid           VARCHAR(240),
 level2Name          VARCHAR(240),
 iucnRedListCategory VARCHAR(240),
 FOREIGN KEY (gbifID) REFERENCES OccurrenceDetails(gbifID)
);
'''
criar_db(sql_occurrence_details)
criar_db(sql_taxonomic_information)
criar_db(sql_species_information)
criar_db(sql_dataset_information)
criar_db(sql_protocol_and_iucn_information)

```

Função para inserir dados no banco

```python
# Função para inserir dados no banco
def inserir_db(sql):
    con = conecta_db()
    cur = con.cursor()
    try:
        cur.execute(sql)
        con.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        con.rollback()
        cur.close()
        return 1
    cur.close()
```

Realizando inserts, trocando valores vazios por null, nas tabelas.

```python
df = dfs['occurrencedetails']

def format_value(value):
    if pd.isnull(value) or value == 'nan':
        return 'NULL'
    elif isinstance(value, str):
        return "'{}'".format(value)
    else:
        return str(value)

for i in df.index:
    values = [format_value(str(df[col][i])) for col in df.columns]
    insert_sql_occurrence_details = """
    INSERT into occurrencedetails ({}) 
    values({});
    """.format(', '.join(df.columns), ', '.join(values))
    inserir_db(insert_sql_occurrence_details)
```
